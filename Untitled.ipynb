```python
#Analyzing de PROD_NAME attribute

words = (
    df_transactions["PROD_NAME"]
    .dropna()
    .str.lower()
    .unique()
)

words = pd.Series(words).str.split().explode()
```

```python
words = words[~words.str.contains(r"\d")]
words = words[words.str.contains(r"[a-z]+$")]
```

```python
words_counts = words.value_counts().sort_values(ascending = False)
print(words_counts.head(15))
```
